```python
import pandas as pd
import requests
import json
import csv
import time
import datetime
start  = 1453680000
end = 1516882331
```

```python
def getPushshiftData(query, after, before, sub):
    url = 'https://api.pushshift.io/reddit/search/submission/?title='+str(query)+'&size=1000&after='+str(after)+'&before='+str(before)+'&subreddit='+str(sub)
    print(url)
    r = requests.get(url)
    data = json.loads(r.text)
    return data['data']
```

```python
data = []
length = 0;
while start <= end:
    temp_end = min(start+150000, end)
    x = getPushshiftData("", start, temp_end, 'india')
    data = data+x
    start = start+150000
len(data)
```
